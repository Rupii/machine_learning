```python
import pandas as pd
from  sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_fwf("body_weight.txt")
x = df[["Brain"]]
y =  df[["Body"]]
```

```python
body_reg = linear_model.LinearRegression()
body_reg.fit(x,y)
```

```python
plt.scatter(x,y)
plt.plot(x,body_reg.predict(x))
plt.show()
```
