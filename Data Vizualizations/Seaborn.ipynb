```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import pandas as pd
import numpy as np
plt.style.use('ggplot')
##ggplot and seaborn both can be use in style use , it's just a matter of preferance
```

** Seaborn comes with few datasets to work upon**

if you are loading the datasets for first time make sure you have a working internet connection

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Distribution plot
# to quickly analyse the total bill
# a dist is univaient plot

sns.distplot(tips['total_bill'], bins = 20, kde = False)

#If don't understand what is kde don't worry, it's just a density estimation (you no need to worry about that).
#to check about it turn it to default
```

```python
#Joint plot
#This is bivarent distribution plot with kind(allows to choose camparison plot)
sns.jointplot(tips['total_bill'], tips['tip'], kind='scatter' )

#use kind "reg" to draw a regression line over the plot
```

```python
#Pair plot
sns.pairplot(data = tips, hue = 'sex', palette = 'gist_rainbow_r')
```

```python
#Rug plot(bivarient)

sns.rugplot(tips['total_bill'])
```

## Catogorical plots

```python
#Box Plot
#x is catagorical y is numeric
# Estimator is use defined aggregate(by defgault it is mean)

sns.barplot(tips['sex'], tips['total_bill'], estimator=np.std)
```

```python
#Count plot
#only x value since y is default count
sns.countplot(tips['sex'])
```

```python
#Box Plot
sns.boxplot(data= tips, x = 'day', y= 'total_bill', hue = 'smoker')
```

```python
#viloin plot
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True )
```

```python
#Strip plot
sns.stripplot(data= tips, x = 'day', y= 'total_bill', jitter = True, hue = 'smoker', split = True)
```

```python
#Swarmplot
sns.swarmplot(data= tips, x = 'day', y= 'total_bill', color = 'black')
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True )
#Does not scale for large datasets, it's plotting all the points the kernel may crash
```

#### Factor Plot
** General plot with Kind**

```python
sns.factorplot(x = 'day', y = 'total_bill', data = tips, kind ='violin')
```

## Matrix Plot

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
sns.heatmap(tips.corr(), annot = True)
```

```python
df = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
sns.heatmap(df, cmap = 'magma', linewidth = 2, linecolor = 'black')
```

```python
sns.clustermap(df)
```
