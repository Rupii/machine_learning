## part I 

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = np.arange(1, 10)
y = x ** 2
```

```python
plt.scatter(x, y)
```

## Functional method

```python
x = np.linspace(1, 6, 11)
y = x ** 2
plt.xlabel('X Label')
plt.ylabel("Y Label")
plt.title("X v Y")
plt.plot(x, y)
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r-', marker = 'h')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

## Objecct Oreinted method|

```python
fig = plt.figure()
axis = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis.plot(x, y)
axis.set_xlabel('X Label')
axis.set_ylabel("Y Label")
axis.set_title("OO plot")
```

```python
#multiple axis
fig = plt.figure()

axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axis2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])
axis1.plot(x, y, 'r-')
axis1.set_title("Larger plot")
axis2.set_title('Smaller')
axis2.plot(y, x, marker = 'd')
```

## part II

```python
fig, axes = plt.subplots(1, 2)
#axes is list viz> iterable
axes[1].plot(y, x)
axes[0].plot(x, y)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 4), dpi=100)
axes[0].plot(x, y  , color = '#ffc800')
axes[0].set_title("X vs Y")
axes[1].set_title("Y vs X")
axes[1].plot(y, x, color = '#d20fae', label = "two")
axes[1].plot(y, x **2, label = "one")
axes[1].legend(loc = 0)

plt.tight_layout()
```

## Part III

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1 ,1])

x = np.linspace(0, 10, 10)

y = x ** 2

ax.plot(x, y, linewidth = 2, alpha= 0.9, marker = 's', linestyle  = '--', color = '#ff000f')
ax.set_xlim([0,100])
ax.set_ylim([0,100
            ])
```
