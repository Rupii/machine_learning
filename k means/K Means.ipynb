# K Means Clusterning

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# make artificial data
from sklearn.datasets import make_blobs
data = make_blobs(n_samples= 200, n_features=2, centers = 4, cluster_std=1.8, 
                  random_state=101)
```

```python
plt.style.use("ggplot")
plt.scatter(data[0][: , 0], data[0][:, 1], c = data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters= 4)
kmeans.fit(data[0])
```

```python
kmeans.labels_
#in real dataset we cannot compare the result
#As we don't exactly know which forms a cluster
#the Whole point of using kmeans is that
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize = (10, 6))

ax1.set_title("K Means")
ax1.scatter(data[0][:, 0], data[0][:,1], c = kmeans.labels_)


ax2.set_title("Original")
ax2.scatter(data[0][:, 0], data[0][:,1], c = data[1])
```
